```python
#Importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

#Reading datafile using pandas
dataset = pd.read_csv("E:\\ML Zero to Hero\\50_Startups.csv")

#Separating files for x and y axis values
x = dataset[["R&D Spend", "Administration", "Marketing Spend", "State"]].values
y = dataset[["Profit"]].values

#Encoding the string valued column ie 'State'
state = ColumnTransformer([("State", OneHotEncoder(), [3])], remainder='passthrough')
x = state.fit_transform(x)

#Splitting dataset into training and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)

#Setting model
LR = LinearRegression()
LR.fit(X_train, Y_train)

#Predicting from the model using test data set
y_pred = LR.predict(X_test)

#Finding Accuracy
acc = (y_pred.sum().mean() / Y_test.sum().mean())

#Printing the accuracy
print("Accuracy of model = " + str(round(acc * 100)) + "%")

#Visualizing the relation between dependent and independent columns

#separating each column for x
x1 = dataset[["R&D Spend"]].values

x2 = dataset[["Administration"]].values

x3 = dataset[["Marketing Spend"]].values

state = ColumnTransformer([("State", OneHotEncoder(), [3])], remainder='passthrough')
x4 = state.fit_transform(x)

y = dataset[["Profit"]].values


#Visualize X1 and y
plt.scatter(x1,y, color='orange')
plt.title("R&D Spend VS Profit")
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.show()

#Visualize X2 and y
plt.scatter(x2,y, color='orange')
plt.title("Administration VS Profit")
plt.xlabel("Administration")
plt.ylabel("Profit")
plt.show()

#Visualize X3 and y
plt.scatter(x3,y, color='orange')
plt.title("Marketing Spend VS Profit")
plt.xlabel("Marketing Spend")
plt.ylabel("Profit")
plt.show()
```
