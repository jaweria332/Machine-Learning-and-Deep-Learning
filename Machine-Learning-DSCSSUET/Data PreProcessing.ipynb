# NORMALIZE DATA

```python
#importing numpy, pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplot

#Generating normalized data using panda
dataset = pd.DataFrame({
    'x1':np.random.normal(3,5,5000),
    'x2':np.random.normal(1,3,5000),
    'x3':np.random.normal(0,10,5000)
})

#Plot data
dataset.plot.kde()
```

# USING STANDARD SCALE TO STANDARDIZE DATA

```python
#importing StandardScaler from sklearn
from sklearn.preprocessing import StandardScaler

#saving the module in a variable
SS = StandardScaler()

#Standardizing the above declared data using StandardScaler
SS_data = SS.fit_transform(dataset)

#converting standardized dataset into columns
dataset = pd.DataFrame(SS_data, columns = ['C1', 'C2', 'C3'])

#visualize the standardized data
dataset.plot.kde()
```

# MINMAX SCALAR

```python
#Generating data that follows chi-squared, beta and normal distribution
data = pd.DataFrame({
    'x1': np.random.chisquare(8,1000),
    'x2': np.random.beta(8, 2, 1000) *40,
    'x3': np.random.normal(50,3,1000)
})

#Plot the data
data.plot.kde()


#Importing minmax scaler
from sklearn.preprocessing import MinMaxScaler
MM = MinMaxScaler()

#transforming data
dataset = MM.fit_transform(data)

#converting transformed dataset into columns
data = pd.DataFrame(dataset, columns = ['x1', 'x2', 'x3'])
data.plot.kde()
```

# ROBUST SCALER

```python
#Declaring data with lower and higher outlier 
data = pd.DataFrame({
    # Distribution with lower outliers
    'Data1': np.concatenate([np.random.normal(20, 1, 1000), np.random.normal(1, 1, 25)]),
    # Distribution with higher outliers
    'Data2': np.concatenate([np.random.normal(30, 1, 1000), np.random.normal(50, 1, 25)]),
})

#Plot the data
data.plot.kde()

#Importing Robust scaler
from sklearn.preprocessing import RobustScaler
RS = RobustScaler()

#Scaling data
dataset=RS.fit_transform(data)

#converting transformed dataset into columns
data = pd.DataFrame(dataset, columns = ['Data1', 'Data2'])

#Plot scaled data
data.plot.kde()
```

# NORMALIZER FOR THREE-DIMENSIONAL DATA

```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #This is important for 3d plotting 


#Declaring three-dimensional data
data = pd.DataFrame({
    'x1': np.random.randint(-100, 100, 1000).astype(float),
    'y1': np.random.randint(-80, 80, 1000).astype(float),
    'z1': np.random.randint(-150, 150, 1000).astype(float),
})

#Plotting figure in 3D
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(data.x1, data.y1, data.z1)
```

```python
#Importing Normalizer
from sklearn.preprocessing import Normalizer
norm = Normalizer()

#Transforming data
dataset = norm.fit_transform(data)

#Drawing data in columns
data = pd.DataFrame(dataset, columns = ['x1','y1','z1'])

#Plotting Normalized Data
ax = plt.axes(projection='3d')
ax.scatter3D(data.x1, data.y1, data.z1)
```

# ENCODING CATEGORICAL VALUES

```python
#Declaring dataset
data = pd.DataFrame({
    'Age': [21,22,19,20,19,19,22,21],
    'Gener': ['Teacher','Lecturer', 'Lecturer', 'Teacher','Teacher','Rechecker', 'Lecturer','Rechecker'],
})

#Categorizing Income attribute using map function
data.Gener.map({'Teacher' : 1, 'Lecturer' : 2, 'Rechecker': 3})
```

# using LabelEncoder and OneHotEncoder

```python
#Import modules
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

#Transform data using module
LE = LabelEncoder()
data['Gener Code'] = LE.fit_transform(data.Gener)

#Check how LabelEncoder encoded data
print(data)


OHE = OneHotEncoder()
OHE.fit_transform(data[['Gener Code']]).toarray()
```

# IMPUTATION TO DEAL WITH NULL VALUES

```python
#Defining dataset
data = pd.DataFrame({
    'ODD': [9,7,5,np.nan,3,1],
    'EVEN':[0,2,4,6,np.nan, 10],
})

#Importing module
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan, strategy = 'mean' )

#Transforming dataset
imp.fit_transform(data)
```
