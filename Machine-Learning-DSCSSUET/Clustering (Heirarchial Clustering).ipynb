# HIERARCHICAL CLUSTERING

Hierarchical clustering combine similar object together in a group. Each of these groups called clusters.
Hierarchical clustering is also known as ***hierarchical cluster analysis***

```python
#Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.datasets.samples_generator import make_blobs



#Loading dataset
x, y = make_blobs(n_samples = 50)
x1 = x[:, 0]
x2 = x[:, 1]

#Visualizing scatter plot
plt.scatter(x1, x2, s=5)
plt.show()


#Implementing nearest point algorithm
linkage_matrix = linkage(x, "complete")
print(linkage_matrix)

#Visualizing tree like structure with truncate mode
dendrogram = dendrogram(linkage_matrix, truncate_mode=None)
plt.title("Hierarchical Clusterring")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
