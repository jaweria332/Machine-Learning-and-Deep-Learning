```python
#Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import math

#Reading dataset
dataset = pd.read_csv("E:\\ML Zero to Hero\\Drugs_Performance.csv")

#Separating columns for x and y axis data
x = dataset[["Age"]].values
y = dataset[["Drugs_Perf"]].values

#Visualizing scatter plot
plt.scatter(x,y)
plt.title("Age VS Drugs Performance")
plt.xlabel("Age")
plt.ylabel("Drugs Performance")
plt.show()

#SCATTER PLOT SIMPLY SHOWS QUADARATIC REGRESSION MODEL FITS BETTER ON DATA

#Separating test and training dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

#Fitting polynomial regression of degree 2 on dataset
pol = PolynomialFeatures(degree = 2)
x_poly = pol.fit_transform(x_train)
pol.fit(x_poly, y_train)

#Train the model
model = LinearRegression()
model.fit(x_poly, y_train)

#Visualizing the regression
plt.scatter(x,y, color='yellow')
plt.plot(x,model.predict(pol.fit_transform(x)), color='purple')
plt.title("Age VS Drugs Performance")
plt.xlabel("Age")
plt.ylabel("Drugs Performance")
plt.show()

#Draw prediction from model
age = int(input("Enter age : "))
DP = model.predict(pol.fit_transform(x_test))
print("Drug Performance = " + str(DP))
```
