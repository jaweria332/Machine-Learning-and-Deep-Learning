# INTRODUCTION TO NEURAL NETWORKS

The defined Neural Network consist of single input neuron, single hidden layer neuron which perform a very basic activation function and generate the output as requirement.

```python
#Importing necessary libraries
import tensorflow as tf
import numpy as np
from tensorflow import keras

#Defining neural network
model=tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

#Compiling neural networks
model.compile(optimizer='sgd', loss='mean_squared_error')

#Providing manual data
x = np.array([-1.0, 0.0, 1.0,2.0,3.0,4.0], dtype=float)
y = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)

#Training the neural network
model.fit(x, y, epochs=500)

#Drawing predictions
print(model.predict([10.0]))
```
