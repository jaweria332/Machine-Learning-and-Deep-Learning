```python
#Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import math

#Read dataset
dataset = pd.read_csv('E:\\ML Zero to Hero\\headbrain.csv')

#separating the data for x and y axis from the given CSV file
x = dataset['Head Size(cm^3)']
y = dataset['Brain Weight(grams)']

#printing out extracted data for self check, can comment these lines
print(x)
print(y)

#Drawing scatter plot
plt.scatter(x,y)
plt.title("Head Size VS Brain Weight")
plt.xlabel("Head Size(cm3)")
plt.ylabel("Brain Weight(grams)")
plt.show()

#Convert dataframes into array because Machine can only handle array, not dataframes
X = np.array(x).reshape(-1,1)
Y = np.array(y).reshape(-1,1)

#Train the model using module
LR = LinearRegression()
LR.fit(X, Y)

#Visualizing the Linear Model
plt.scatter(x,y, color='pink')
plt.plot(X, LR.predict(X), color='purple')
plt.title("Head Size VS Brain Weight")
plt.xlabel("Head Size(cm3)")
plt.ylabel("Brain Weight(grams)")
plt.show()

#Finding Accuracy and Error of model
MSE = mean_squared_error(X, Y)
print("\nMean Squared Error : ", math.sqrt(MSE))
print("\nAccuracy (Score) : " + str(round((LR.score(X, Y)) * 100))+ " %")

#Predict brainweight according to the user input
HeadSize = int(input("\nEnter Head Size : "))
BrainWeight = LR.predict([[HeadSize]])
print("\nBrain Weight (Calculated by model) = ", BrainWeight)
```
